### Load the ‘GSSvocab’dataset from the ‘car’package. This dataset contains people’s scores on an Englishvocabulary test and includes demographic information.


```r
# This is data from Alina, because there was a problem with the car package.

load('GSSvocab.RData')
str(GSSvocab)
```

```r
GSSvocab
```

### Filter for the year 1978 and remove rows with missing values (the function na.exclude() is one way to do this - check out the documentation!). 

```r
# This code filters for the year 1978, and removes missing values.

dat1978 <- GSSvocab %>%
    filter(year==1978)
    na.exclude(dat1978)
str(dat1978)
```

### Is a person’s  score on the vocabulary test  (‘vocab’)significantly impacted by their level  of education (‘educ’)? Visualize the relationship in a plot and build a model. Briefly explain the results.

```r
# The plot for vocabulary and education.

ggplot(dat1978,
       aes(x = educ,
           y = vocab)) +
    geom_point(position = 'jitter') +
        geom_point()
```

```r
cor.test(dat1978$vocab, dat1978$educ)

# The correlation coefficient indicates an okay correlation between vocabulary and education. 
# - The correlation isn't perfect - but as education increases, vocabulary scores also increase.

# The p-value is very small, which suggests that the correlation is statistically significant.
```

```r
vocab_educ_model <- lm(vocab ~ educ, dat1978)
summary(vocab_educ_model)

# Under this, is the model for vocabulary and years of education.


# The vocabulary score goes up by 0.394 for each increase in years of education. 
# The p-value for this is low, which indicates that it's a statistically significant effect, 
# - and that there's a correlation between education and vocabulary. 

# However, R-squared is not that high, which indicates that other factors than education 
# - are probably involved in predicting vocabulary.
```

### Whether a person is the native of an English-speaking country(‘nativeBorn’) could potentially have an impact on the size of their vocabulary. Visualize the relationship and add the predictor to the model. Briefly explain the results.

```r
ggplot(dat1978,
       aes(x = nativeBorn,
           y = vocab)) +
    geom_point(position = 'jitter') +
        geom_point()

# The plot for vocabulary and native of an English-speaking country.
```

```r
vocab_educ_nativeBorn_model <- lm(vocab ~ educ + nativeBorn, dat1978)
summary(vocab_educ_nativeBorn_model)

# Under this, is the model for vocabulary, years of education and native of an English-speaking country.


# The vocabulary score goes up by 0.65 when the person is a native of an English-speaking country, when controlling for education. 
# The p-value for this is low, which indicates that it's statistically significant. 

# In this model, R-squared is 0.2926.
# This is better than before, which indicates that adding the "nativeBorn" variable is good.
```

### Does  a  person’s  level  of  education  depend  on  whether they  are  a  native  of  the  country? Visualize the relationship. Do you think it makes sense to add the relationship as an interaction term? Try creating the model and briefly explain the results.

```r
ggplot(dat1978,
       aes(x = nativeBorn,
           y = educ)) +
    geom_point(position = 'jitter') +
        geom_point()

# The plot for education and native of an English-speaking country.
```

```r
vocab_educ_nativeBorn_model_2 <- lm(vocab ~ educ * nativeBorn, dat1978)
summary(vocab_educ_nativeBorn_model_2)

# The model for vocabulary, education and native of an English-speaking country.


# The coefficient for the interaction term is -0.025, which means that (for native-born individuals), 
# - each year of education adds a bit less to the vocabulary score, when they're compared to the non-natives. 
# The p-value is low, indicating that the effect is not statistically significant.

# In this model, R-squared is 0.2926 again.
# This is the same as before, which indicates that adding "nativeBorn" 
# - as an interaction term actually doesn't improve the fit of the model. 

```

### Which model performs best?

```r
# I will test which model performs best, by making an anova test.


# I found this code, which helped streamline the data. I'm not quite sure what the root of the problem was, 
# but this piece of code helped fix it.
dat1978_v2 <- dat1978[complete.cases(dat1978[, c("vocab", "educ", "nativeBorn")]), ]


# I then coded the anova test, as we have learned. 
m1 <- lm(vocab ~ educ, dat1978_v2)
m2 <- lm(vocab ~ educ + nativeBorn, dat1978_v2)
m3 <- lm(vocab ~ educ * nativeBorn, dat1978_v2)

anova(m1, m2, m3)
```

```r
# The anova test shows that model 2 was indeed the best option.
# Adding the nativeBorn as a variable increases the "Sum of Sq" by a lot, which is good. 
# Adding the interaction term in model 3 does not do much good, with only a slight increase in "Sum of Sq".

# This is a nice condensed version of the information from the earlier tests.


# Overall, I finally feel like I have some idea of the statistics part of this course. 
# It's still difficult, but easier than at the start.
# I was in the valley of despair for while, but with a lot of repetetion and a lot of help, it all worked out :)
```
