## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata_1 = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
# I start by making a dataframe for each decade, which i then combine into one dataframe.

df_0 = rawdata_1[0]
df_1 = rawdata_1[1]
df_2 = rawdata_1[2]
df_3 = rawdata_1[3]
df_4 = rawdata_1[4]
df_5 = rawdata_1[5]
df_6 = rawdata_1[6]
df_7 = rawdata_1[7]

df_space = pd.concat([df_0, df_1, df_2, df_3, df_4, df_5, df_6, df_7], ignore_index=True, sort=False)
df_space
```

```python
# I was a bit stumped on extracting the year, so I asked our friendly "co-pilot-ish" for help.
# The help was scarily good, but not without flaws. The code started by only outputting 1965, but when I checked the data manually, 1967 had the same number of launches.
# The help is nice, but i make sure to look through the code, so i can actually understand what is happening:)


# First, I have to extract the year som the column containing the launch dates.
df_space['Year'] = pd.to_datetime(df_space['Launch date']).dt.year

# Then I will have to use 'groupby' to group by year and count the launches for each year.
launches_per_year = df_space.groupby('Year').size()

# Then I extract the maximum number of launches.
max_launches = launches_per_year.max()

# Lastly, I find all the years with maximum launches, and print the result with some text.
years_with_max_launches = launches_per_year[launches_per_year == max_launches].index.tolist()

print(f"The maximum number of launches are {max_launches}. The years with the most launches are: {years_with_max_launches}.")
```

```python
# I use the same bits of code, and modify them a bit, to find the month with the most launches.
# The month with the most launches is '11' which corresponds to November.

df_space['Month'] = pd.to_datetime(df_space['Launch date']).dt.month
launches_per_month = df_space.groupby('Month').size()
max_launches_month = launches_per_month.max()
months_with_max_launches = launches_per_month[launches_per_month == max_launches_month].index.tolist()

print(f"The maximum number of launches per month are {max_launches_month}. The month with the most launches are: {months_with_max_launches}.")
```

```python
# I use this bit of code to sort the values in 'launches_per_month' which is a panda.series.Series

sorted_launches = launches_per_month.sort_values(ascending=False)
sorted_launches
```

2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata_2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
## 1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's


# Same concept as before, this time with villians

df_00 = rawdata_2[3]
df_01 = rawdata_2[4]
df_02 = rawdata_2[5]
df_03 = rawdata_2[6]
df_04 = rawdata_2[7]
df_05 = rawdata_2[8]
df_06 = rawdata_2[9]
df_07 = rawdata_2[10]
df_08 = rawdata_2[11]

df_villain = pd.concat([df_00, df_01, df_02, df_03, df_04, df_05, df_06, df_07, df_08], ignore_index=True, sort=False)
df_villain
```

```python
## 2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade

# Start by converting the data in "Year Debuted" to decades.
df_villain['Year Debuted'] = pd.to_numeric(df_villain['Year Debuted'], errors='coerce')

df_villain['Decade'] = (df_villain['Year Debuted'] // 10) * 10

# As with the space challenge, we use "groupby" to group the decades together and find the number of debuts in each decade.
decade_count = df_villain.groupby('Decade').size()

# Lastly we sort the data, thereby ranking them
decade_count_sort = decade_count.sort_values(ascending=False)

decade_count_sort
```
